```python
import pandas as pd
df = pd.read_csv("characterization_dataset.csv")

print("First 5 Rows:\n", df.head())
print("\nInfo:\n")
print(df.info())

print("\nSummary:\n")
print(df.describe(include='all'))
```

```python
# step 2 : Check missing values count 
print("\nMissing Values Count:\n", df.isnull().sum())

missing_percentage = df.isnull().mean() * 100
print("\nMissing %:\n", missing_percentage)
```

```python
#Step 1: Inspect Data Types & Summary 
# Column types 
print(df.dtypes) 
# Descriptive statistics 
print(df.describe(include='all'))
```

```python
#Step 3: Explore Data from Multiple Sources 
# Example: Combine multiple CSV files 
df1 = pd.read_csv('Heart.csv') 
df2 = pd.read_csv('StudentsPerformance.csv') 
# Combine vertically 
df_combined = pd.concat([df1, df2], ignore_index=True) 
# Combine horizontally (on key) 
df_merged = pd.merge(df1, df2, on='gender', how='inner') 
print("Combine vertically : \n",df_combined)
print("\nCombine horizontally : \n",df_merged)
```

```python
#Step 4: Visualize Missing Patterns 
import matplotlib.pyplot as plt
import missingno as msno 
import seaborn as sns
# Matrix plot 
msno.matrix(df) 
plt.show() 
# Bar chart 
msno.bar(df) 
plt.show()
```

```python
#Step 5: Summary Statistics for Characterization 
# Categorical column distribution 
for col in df.select_dtypes(include='object'): 
    print(df[col].value_counts(normalize=True) * 100) 
# Correlation matrix for numerical columns 
print(df.corr(numeric_only=True))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```
