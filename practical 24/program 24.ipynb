```python
import pandas as pd


# STEP 1: Load Dataset

df = pd.read_csv('lab_data.csv',parse_dates=['ts'])

print("Original Data:")
print(df.head())


# STEP 2: Set Timestamp Index

df = df.set_index('ts')

# Sort timestamps
df = df.sort_index()

print("\nIndexed Data:")
print(df.head())


# STEP 3: Resampling

# Convert data into 15-minute intervals
df_prepared = df.resample('15T').mean()

print("\nAfter Resampling:")
print(df_prepared.head(10))


# STEP 4: Interpolation

# Fill missing values
df_prepared['value'] = df_prepared['value'].interpolate(method='time')

print("\nAfter Interpolation:")
print(df_prepared.head(10))


# STEP 5: Save Prepared Data

df_prepared.to_csv('prepared_lab_data.csv')

print("\nPrepared dataset saved successfully.")
```
