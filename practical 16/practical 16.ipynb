```python
import pandas as pd
import json

# STEP 1: Load JSON data from file
with open('data.json', 'r') as f:
    data = json.load(f)

# STEP 2: Manually Flatten JSON (NO json_normalize)
flat_data = []

for item in data:
    new_row = {
        "id": item["id"],
        "name": item["info"]["name"],
        "price": item["info"]["price"],
        "tags": item["tags"]  # tags is now a single string, avoiding duplicates
    }
    flat_data.append(new_row)

df_final = pd.DataFrame(flat_data)

print("=== Step 3: Flattened Data ===")
print(df_final)

# STEP 4: Summarization
print("\n=== Statistics (Price) ===")
print(df_final["price"].describe())

# STEP 5: Export CSV
df_final.to_csv("structured_output.csv", index=False)

print("\nFile 'structured_output.csv' created successfully!")
```
