```python
# DATA ACCESS CONTROL WITH HTML TABLE DISPLAY

import pandas as pd
from IPython.display import display, HTML

# STEP 1: Load dataset
data = pd.read_csv("student_performance_data.csv")

# STEP 2: Define allowed columns
allowed_columns = ['StudentID', 'Gender', 'Age', 'Major']

# STEP 3: Create restricted dataset
restricted_data = data[allowed_columns]

# STEP 4: Display restricted dataset in table format
print("Restricted Dataset (Authorized View):")
display(HTML(restricted_data.head().to_html()))
```

```python
# DATA PRIVACY PROTECTION WITH HTML TABLE DISPLAY

import pandas as pd
from IPython.display import display, HTML

# Load dataset
data = pd.read_csv("student_performance_data.csv")

# STEP 1: Create copy (protect original data)
privacy_data = data.copy()

# STEP 2: Mask StudentID
privacy_data['StudentID'] = privacy_data['StudentID'].apply(
    lambda x: "ID_***" + str(x)[-3:]
)

# STEP 3: Display anonymized dataset in table format
print("Anonymized Dataset:")
display(HTML(privacy_data.head().to_html()))
```
