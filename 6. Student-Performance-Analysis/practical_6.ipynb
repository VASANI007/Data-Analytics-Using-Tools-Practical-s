```python

#  DATA STORAGE WITH HTML TABLE OUTPUT (DISPLAY)


import pandas as pd
from IPython.display import display, HTML

# STEP 1: Load dataset
data = pd.read_csv("student_performance_data.csv")

# STEP 2: Display First 5 Rows
print("First 5 Rows:")
display(HTML(data.head().to_html()))

# STEP 3: Display Dataset Shape
print("\nDataset Shape:")
shape_df = pd.DataFrame({
    "Rows": [data.shape[0]],
    "Columns": [data.shape[1]]
})
display(HTML(shape_df.to_html(index=False)))

# STEP 4: Display Dataset Info
print("\nDataset Info:")
info_df = pd.DataFrame({
    "Column": data.columns,
    "Non-Null Count": data.count().values,
    "Data Type": data.dtypes.values
})
display(HTML(info_df.to_html(index=False)))

# STEP 5: Save dataset copy
data.to_csv("student_data_storage_copy.csv", index=False)

print("\nDataset successfully stored and accessed.")
```

```python

# DATA INTEGRATION WITH HTML TABLE OUTPUT (DISPLAY)


import pandas as pd
from IPython.display import display, HTML

# STEP 1: Load dataset
data = pd.read_csv("student_performance_data.csv")

# STEP 2: Create two datasets
academic_data = data[['StudentID','GPA','Major']]
activity_data = data[['StudentID','ExtraCurricularActivities','PartTimeJob']]

# STEP 3: Display Academic Data
print("Academic Data Sample:")
display(HTML(academic_data.head().to_html()))

# STEP 4: Display Activity Data
print("\nActivity Data Sample:")
display(HTML(activity_data.head().to_html()))

# STEP 5: Merge datasets
integrated_data = pd.merge(
    academic_data,
    activity_data,
    on="StudentID"
)

# STEP 6: Display Integrated Data
print("\nIntegrated Dataset:")
display(HTML(integrated_data.head().to_html()))
```

```python

# PATTERN DISCOVERY WITH HTML TABLE OUTPUT (DISPLAY)


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# STEP 1: Load dataset
data = pd.read_csv("student_performance_data.csv")

# STEP 2: Statistical Summary (HTML Table)
print("Statistical Summary:")
display(HTML(data.describe().to_html()))

# STEP 3: GPA Distribution
plt.hist(data['GPA'], bins=10)
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.show()

# STEP 4: Study Hours vs GPA
plt.scatter(data['StudyHoursPerWeek'], data['GPA'])
plt.title("Study Hours vs GPA")
plt.xlabel("Study Hours Per Week")
plt.ylabel("GPA")
plt.show()

# STEP 5: Attendance vs GPA
plt.scatter(data['AttendanceRate'], data['GPA'])
plt.title("Attendance Rate vs GPA")
plt.xlabel("Attendance Rate")
plt.ylabel("GPA")
plt.show()

# STEP 6: Correlation Matrix (FIXED for all pandas versions)
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()

print("\nCorrelation Matrix:")
display(HTML(correlation_matrix.to_html()))

# STEP 7: Heatmap Visualization
sns.heatmap(correlation_matrix, annot=True)
plt.title("Feature Correlation Heatmap\n")
plt.show()
```
