```python
import pandas as pd
import matplotlib.pyplot as plt# 1. Load Tabular Data
# Use pd.read_csv("your_file.csv") for external files
df = pd.read_csv("scottish_hills.csv")
# 2. Structured Summarization
print("--- Dataset Snapshot ---")
print(df.head(10)) # First 10 rows for initial interpretation
print("\n--- Summary Statistics ---")
# Generates mean, median, min, max, and std deviation
print(df.describe())
# 3. Categorical Interpretation
# Example: Grouping by a category to see trends (e.g., Average height per Region)
if 'Region' in df.columns:
 print("\n--- Region-wise Summary ---")
 print(df.groupby('Region')['Height'].mean())
# 4. Visual Interpretation (Box Plot)
# Useful for seeing distributions and identifying outliers
df.boxplot(column='Height')
plt.title('Distribution of Heights')
plt.ylabel('Meters')
plt.show()
```
