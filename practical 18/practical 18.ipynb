```python
import pandas as pd
import numpy as np

# STEP 1: Create Dataset

df = pd.read_csv("data.csv")

print("--- Step 1: Basic Dataset ---")
print(df)

# STEP 2: Enhancement (Feature Engineering)
df['Profit'] = df['Revenue'] - df['Operating_Costs']
df['Margin_%'] = (df['Profit'] / df['Revenue'] * 100)

print("\n--- Step 2: Enhanced Data ---")
print(df[['Branch', 'Revenue', 'Profit', 'Margin_%']])

# STEP 3: Scaling (Min-Max Normalization)
min_rev = df['Revenue'].min()
max_rev = df['Revenue'].max()

df['Rev_Score'] = (df['Revenue'] - min_rev) / (max_rev - min_rev)

print("\n--- Step 3: Normalized Revenue ---")
print(df[['Branch', 'Revenue', 'Rev_Score']])

# STEP 4: Pivot Table (Restructuring)
pivot_df = df.pivot_table(index='Branch', columns='Month', values='Profit', aggfunc='mean') 


print("\n--- Step 4: Pivot Table (Profit by Branch & Month) ---")
print(pivot_df)

# STEP 5: Aggregation Summary
summary = df.groupby('Branch').agg({
 'Revenue': 'sum',
 'Profit': 'sum',
 'Margin_%': 'mean'
})

summary = summary.rename(columns={'Margin_%': 'Avg_Margin'})

print("\n--- Step 5: Final Summary ---")
print(summary)
```
